## Backpropagation Algorithm

Backpropagation is neural network terminology for minimizing our cost function like gradient descent in logistics and linear regression.


$ g^{'}(z^{(3)}) = a^{(3)} .\times (1-a^{(3)}) $, here 1 is a vector with values 1 


Note: there is no $ \delta ^ {1} $ as we dont associate input terms with error.

$ \delta ^ {4} $ error values for the last layer $ \delta ^ {1} $ are simply the differences of our actual results in the last layer and the correct outputs in the y.

$ \delta ^{(l)} = ((\Theta ^ {(l)})^ {T} \delta ^ {(l+1)}) .* a^{(l)} .\times (1-a^{(l)}) $

Here $g ^ {'}$ (g-prime) is the derivative of the activation function g evaluated with the input values given by $ z ^ {(l)} $


The capital-delta matrix D is used as an "accumulator" to add up our values as we go along and eventually compute our partial derivative.
